```python
image = cv2.imread("cat.jpg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

Horizontal Flip

```python
flip_image = cv2.flip(image_rgb, 1)

plt.imshow(flip_image)
plt.axis("off")
plt.title("Flipped Image")
plt.show()
```

Rotation

```python
height, width = image_rgb.shape[:2]

matrix = cv2.getRotationMatrix2D((width/2, height/2), 45, 1)

rotated = cv2.warpAffine(image_rgb, matrix, (width, height))

plt.imshow(rotated)
plt.axis("off")
plt.title("Rotated Image")
plt.show()
```

Brightness Change

```python
bright = cv2.convertScaleAbs(image_rgb, alpha=1, beta=50)

plt.imshow(bright)
plt.axis("off")
plt.title("Brightened Image")
plt.show()
```

Save Augmented Images

```python
cv2.imwrite("flip.jpg", cv2.cvtColor(flip_image, cv2.COLOR_RGB2BGR))
cv2.imwrite("rotated.jpg", cv2.cvtColor(rotated, cv2.COLOR_RGB2BGR))
cv2.imwrite("bright.jpg", cv2.cvtColor(bright, cv2.COLOR_RGB2BGR))

print("Augmented images saved.")
```
